```python
# Traffic Data Analysis Notebook
# Author: Taiseer abuilfelat

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('traffic.csv')
df.head()
```

1. Vehicle Distribution Analysis

```python

plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
sns.boxplot(df['CarCount'])
plt.title('Car Count Distribution')

plt.subplot(2, 2, 2)
sns.boxplot(df['BikeCount'])
plt.title('Bike Count Distribution')

plt.subplot(2, 2, 3)
sns.boxplot(df['BusCount'])
plt.title('Bus Count Distribution')

plt.subplot(2, 2, 4)
sns.boxplot(df['TruckCount'])
plt.title('Truck Count Distribution')

plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
sns.histplot(df['CarCount'], kde=True)
plt.title('Car Count Distribution')

plt.subplot(2, 2, 2)
sns.histplot(df['BikeCount'], kde=True)
plt.title('Bike Count Distribution')

plt.subplot(2, 2, 3)
sns.histplot(df['BusCount'], kde=True)
plt.title('Bus Count Distribution')

plt.subplot(2, 2, 4)
sns.histplot(df['TruckCount'], kde=True)
plt.title('Truck Count Distribution')

plt.tight_layout()
plt.show()
```

2. Traffic Situation Distribution

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Traffic Situation')
plt.title('Traffic Situation Distribution')
plt.show()

traffic_counts = df['Traffic Situation'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(traffic_counts, labels=traffic_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Traffic Situation Distribution')
plt.show()
```

3. Variation by Day of the Week

```python
df['Day of the week'] = pd.to_datetime(df['Time']).dt.day_name()

plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Day of the week', y='Total', estimator=sum)
plt.title('Total Vehicle Count by Day of the Week')
plt.show()
```

4. Car Count vs. Traffic Situation

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Traffic Situation', y='CarCount')
plt.title('Car Count vs. Traffic Situation')
plt.show()
```

5. Bike Count vs. Traffic Situation

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Traffic Situation', y='BikeCount')
plt.title('Bike Count vs. Traffic Situation')
plt.show()
```

6. Bus Count vs. Traffic Situation

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Traffic Situation', y='BusCount')
plt.title('Bus Count vs. Traffic Situation')
plt.show()
```

7. Truck Count vs. Traffic Situation

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Traffic Situation', y='TruckCount')
plt.title('Truck Count vs. Traffic Situation')
plt.show()
```

8. Total Vehicle Count and Traffic Situation

```python
total_by_traffic = df.groupby('Traffic Situation')['Total'].sum()

plt.figure(figsize=(10, 6))
sns.barplot(x=total_by_traffic.index, y=total_by_traffic.values)
plt.title('Total Vehicle Count by Traffic Situation')
plt.show()
```

9. Busiest Hours of the Day

```python
df['Hour'] = pd.to_datetime(df['Time']).dt.hour

total_by_hour = df.groupby('Hour')['Total'].sum()


plt.figure(figsize=(12, 6))
sns.barplot(x=total_by_hour.index, y=total_by_hour.values)
plt.title('Total Vehicle Count by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Total Vehicle Count')
plt.show()
```
